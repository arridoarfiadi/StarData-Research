# Introduction

## What is numpy?


*   A python library.
*   Is a multi-dimensional array object where you can easily do mathematical calculations (i.e., mean, mode, etc.) and complex computations (i.e., merging, joining, indexing)

## Why use NumPy?


*   More efficient than regular Python list because the data type is uniformed. And the memory is fixed, meaning if you only had an numpy array size (3,4) it would only allocate block size of 12 where as standard python list doesn't have a fixed size and it allocates more memory than necessary. The NumPy package breaks down a task into multiple fragments and then processes all the fragments parallelly.
*   Easier to perform matrix multiplications and mathematical operations (built-in functions made available)
*   Well maintained by a strong legitimate community. And everyone pretty much uses it in the data community.




## NumPy vs. Standard Python Lists


*   Numpy has to be the same data type, where standard python lists can contain multiiple data types.
*   Must instantiate the size of a numpy array, where as a standard python list can be empty when you first instantiate it
*   Broadcasting; able to perform complex operations much quuicker and easier in numpy than with standard list. See below for examples.
*   Numpy list is immutable, where as a standard python list is mutable. (i.e., when appending to a numpy array, it's actually creating a brand new numpy array object, but you can still change the value of a specific index)



Numpy vs standard list examples

```python
import numpy as np

a = np.array([[1, 2, 3], [4, 5, 6]])
b = np.array([10, 20, 30])


c = a + b

c
```

## Concatenating

All arrays must be same dimensions in order to concatenate. By defauly it will join on the column. Change the axis parameter to decide how to concatenate.

```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
c = np.concatenate((a, b))
print(c)
```

```python
import numpy as np

a = np.array([[1, 2, 3], [10,4,6]])
b = np.array([[4, 5, 6], [10,4,3]])
c = np.concatenate((a, b), axis=1)
print(c)
```

## Statistical Operations


### Mean (avg)

```python
import numpy as np

a = np.array([1, 2, 3, 4, 5, 6])
np.mean(a)
a.mean()
```

### Median (middle value)

```python
import numpy as np

a = np.array([1, 2, 3, 4, 5, 6])
np.median(a)
```

### Mode (most frequent)

```python
import numpy as np
from scipy import stats

a = np.array([[1, 3, 4, 2, 2, 7],
              [5, 2, 2, 1, 4, 1],
              [3, 3, 2, 2, 1, 1]])

stats.mode(a)
```

### Range (min-max)

```python
import numpy as np

a = np.array([1, 2, 3, 4, 5, 6])
np.ptp(a)
```

### Standard Deviation

```python
import numpy as np

a = np.array([1, 2, 3, 4, 5, 6])
np.std(a)
```
